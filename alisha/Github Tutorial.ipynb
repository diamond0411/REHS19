## <a name="github-commandline"></a>Setting Up GitHub Through the Command Line

To set up a your Github account through the command line, first open a terminal window.

Then, follow these steps...

### <a name="install"></a>Install Github:

#### <a name="install-mac"></a>Mac OS:

There are several ways to install git on your Mac. However, probably the easiest way is to install the Xcode Command Line Tools, which you should be able to do by simply typing to run git from your Terminal:

```
(base) Alishas-MacBook-Pro:~ alishachakraborty$ git --version
```

If it's not already installed, you should be prompted to install it.

If the above option does not work or you need a more up-to-date version of git, you can always install it via a binary installer maintained by the git team, which is available for download at:

<https://git-scm.com/download/mac.>

Alternatively, you can install Git and the osxkeychain helper by using [Homebrew](http://brew.sh/):

```
(base) Alishas-MacBook-Pro:~ alishachakraborty$ brew install git
```


#### <a name="install-windows"></a>Windows:

There are also several ways to install git under Windows. The official binary executable is available for download on the git website:

https://git-scm.com/download/win

However, if you've chosen to work on the Windows Subsystem for Linux this week, then simply follow the installation directions for Linux given below.

#### <a name="install-linux"></a>Linux:

If you want to install git on a Linux-based operating system, you should be able to do so via your operating system's standard package management tool. For example, on any RPM-based Linux distribution, such as Fedora, RHEL, or CentOS, you can use dnf:

``` 
(base) Alishas-MacBook-Pro:~ alishachakraborty$ sudo dnf install git-all
```

Or on any Debian-based distribution, such as Ubuntu, try apt:

```
(base) Alishas-MacBook-Pro:~ alishachakraborty$ sudo apt install git-all
```




### <a name="username"></a>Set up your Git username:

```
base) Alishas-MacBook-Pro:~ alishachakraborty$ git config --global user.name "alishachakraborty"
(base) Alishas-MacBook-Pro:~ alishachakraborty$
```

### <a name="confirm-username"></a>Confirm Git username:

```
(base) Alishas-MacBook-Pro:~ alishachakraborty$ git config --global user.name

alishachakraborty

(base) Alishas-MacBook-Pro:~ alishachakraborty$ 
```

Note: If you want to set you Git username for a single repository, change your current working directory to the local repository where you want to configure the name that is associated with your Git commits.

### <a name="initilize"></a>Initilize an existing directory as a Git repository

Create a directory and cd into in. To initilize an exisitng directory as a Git repo, use the command below:

(base) Alishas-MacBook-Pro:~ alishachakraborty$ git init

### <a name="currentstate"></a>Current State of Repository

These commands are most commonly used for local repositories, a repo that is only contributed by one account.
#### git status

```
(base) Alishas-MacBook-Pro:REHS19_directory alishachakraborty$ git status
```

This lists which (unstaged) files have changed

#### git diff

```
base) Alishas-MacBook-Pro:REHS19_directory alishachakraborty$ git diff
```

This lists (unstaged) changes to files

#### git log

```
(base) Alishas-MacBook-Pro:REHS19_directory alishachakraborty$ git log
```

This lists recent commits

### Adding Files to a Repository

#### Stage File

```
git add filename
```

#### Commit File

```
git commit -m "message"
```

or 

```
 git commit -am "message"
```

This adds/commits all changes from all tracked files (no untracked files) in one go

#### Push File

```
git push
```

This pushes file onto the repository

### <a name="remote"></a>Remote Repositories

A remote repo in Git is a common repository that all team members use to exchange their changes. In most cases, such a remote repository is stored on a code hosting service like GitHub or on an internal server.

#### Add a Remote Repository

```
git remote add origin git@example.com:example/petshop.git
```

#### Push a Current Local repo to Remote

```
git push -u origin master
```

Note: -u sets it to default for the future

#### Show the Avaliable Remote Repositories that have been Added 

```
git remote -v show
```

#### Checkout and Merge Remote Changes

```
git pull
```

Note:

* this does both checkout and merge remote changes in one go

* I recommend executing this command on the command line every time the new repository is updated so the new files are uploaded in the directory where the clonned repo is

#### Update the local cache of the remote repository

```
git fetch origin
```

#### Bring remote refs up to date

```
git remote -v update
```

Note: -v show which branches were updated

#### Branch status

```
git status -uno
```

This will tell you whether the branch you are tracking is ahead, behind or has diverged. If it says nothing, the local and remote are the same.

#### Local <-> remote branch tracking and sync status

```
git show remote origin
```

#### Show commit messages

```
git log HEAD..origin/master --oneline
```

#### Show all changes on remote repo

```
git diff HEAD..origin/master
```

### <a name="branch"></a>Branch Commands

 A branch is essentially a unique set of code changes with a unique name. Repos can have many different branches, but the Master branch should not be changed.

#### Current Branches

To list currently existing branches...

```
git branch
```

#### Create a new branch

````
$ git branch [branchname]

```

You can switch to the branch create in terminal by going to Branch:


<img src="./Images/branch.png" alt="branches " width="300px"  align="left"/>

You can also create a new branch from here as well


#### Move to a branch

```
$ git checkout branchname
```

#### Create/ Checkout a new branch

```
$ git checkout -b branchname

```

#### Remove Branch

```
$ git branch -d branchname
````

#### Merge branch

```
$ git merge [branch]
```

This switched to another branch's history into the current one
